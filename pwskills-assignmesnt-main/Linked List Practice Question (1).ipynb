```python
#1- Define a doubly linked list.
```

```python
# Define the Node class
class Node:
    def __init__(self, data):
        self.data = data  
        self.next = None  
        self.prev = None  

# Define the Doubly Linked List class
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the beginning of the list
    def insert_at_beginning(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    # Function to delete a node by its value
    def delete_node(self, key):
        current = self.head
        # Traverse the list to find the node
        while current:
            if current.data == key:
                if current.prev:  # Adjust previous node's next pointer
                    current.prev.next = current.next
                if current.next:  # Adjust next node's prev pointer
                    current.next.prev = current.prev
                if current == self.head:  # Move head if necessary
                    self.head = current.next
                current = None  # Node is deleted
                return
            current = current.next
        print("Node with value", key, "not found.")

    # Function to traverse the list forward
    def traverse_forward(self):
        current = self.head
        while current:
            print(current.data, end=" <-> ")
            current = current.next
        print("None")

    # Function to traverse the list backward
    def traverse_backward(self):
        current = self.head
        if current is None:
            print("List is empty")
            return
        # Traverse to the last node
        while current.next:
            current = current.next
        # Traverse backward from the last node
        while current:
            print(current.data, end=" <-> ")
            current = current.prev
        print("None")

# Example Usage:

# Create a doubly linked list
dll = DoublyLinkedList()

# Insert nodes at the beginning
dll.insert_at_beginning(10)
dll.insert_at_beginning(20)
dll.insert_at_beginning(30)

# Traverse the list forward
print("Traverse Forward:")
dll.traverse_forward()

# Traverse the list backward
print("\nTraverse Backward:")
dll.traverse_backward()

# Insert node at the end
dll.insert_at_end(40)

# Traverse again after inserting at the end
print("\nTraverse Forward After Insertion at End:")
dll.traverse_forward()

# Delete a node
dll.delete_node(20)

# Traverse after deleting a node
print("\nTraverse Forward After Deletion:")
dll.traverse_forward()
```

```python
#2- Write a function to reverse a linked list in-place.
```

```python
def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev     
            prev = current          
            current = next_node     
        self.head = prev
```

```python
#3- Detect cycle in a linked list.
```

```python
def detect_cycle(self):
        slow = self.head
        fast = self.head
        
        while fast and fast.next:
            slow = slow.next    
            fast = fast.next.next
            
            if slow == fast:  
                return True
        return False 
```

```python
#4- Merge two sorted linked list into one.
```

```python
class Node:
    def __init__(self, data):
        self.data = data  
        self.next = None  

class SinglyLinkedList:
    def __init__(self):
        self.head = None  

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    # Function to traverse the list and print its elements
    def traverse(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    # Function to merge two sorted linked lists into one sorted list
    def merge_sorted_lists(self, l1, l2):
        dummy = Node(0)  
        current = dummy
        
        # Traverse through both lists and merge them
        while l1 and l2:
            if l1.data < l2.data:
                current.next = l1
                l1 = l1.next
            else:
                current.next = l2
                l2 = l2.next
            current = current.next
        
        # If any list still has remaining nodes, append them to the merged list
        if l1:
            current.next = l1
        elif l2:
            current.next = l2
        
        # The merged list is next to the dummy node
        return dummy.next


# Example Usage:

# Create two sorted linked lists
l1 = SinglyLinkedList()
l1.insert_at_end(1)
l1.insert_at_end(3)
l1.insert_at_end(5)
l1.insert_at_end(6)

l2 = SinglyLinkedList()
l2.insert_at_end(2)
l2.insert_at_end(4)
l2.insert_at_end(6)
l2.insert_at_end(8)

# Merge the two sorted linked lists
merged_list = SinglyLinkedList()
merged_list.head = merged_list.merge_sorted_lists(l1.head, l2.head)

# Traverse the merged list
print("Merged Sorted List:")
merged_list.traverse()
```

```python
#5- Write a function to remove nth node from the end in a linked list.
```

```python
def remove_nth_from_end(self, n):
        # Create a dummy node
        dummy = Node(0)
        dummy.next = self.head
        fast = dummy
        slow = dummy
        
        # Move fast pointer n+1 steps ahead to maintain the gap
        for _ in range(n + 1):
            if fast:
                fast = fast.next
        
        # Move both fast and slow pointers until fast reaches the end
        while fast:
            fast = fast.next
            slow = slow.next
        
        # Remove the nth node
        slow.next = slow.next.next

        # Update the head in case the head node was removed
        self.head = dummy.next
```

```python
#6- Remove duplicates from a sorted linked list
```

```python
class Node:
    def __init__(self, data):
        self.data = data 
        self.next = None 

class SinglyLinkedList:
    def __init__(self):
        self.head = None 

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    # Function to traverse the list and print its elements
    def traverse(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    # Function to remove duplicates from a sorted linked list
    def remove_duplicates(self):
        current = self.head
        
        # Traverse the list
        while current and current.next:
            # If current node's data is same as next node's data, skip the next node
            if current.data == current.next.data:
                current.next = current.next.next
            else:
                current = current.next


# Example Usage:

# Create a sorted singly linked list with duplicates
sll = SinglyLinkedList()

# Insert nodes
sll.insert_at_end(1)
sll.insert_at_end(2)
sll.insert_at_end(3)
sll.insert_at_end(3)
sll.insert_at_end(4)
sll.insert_at_end(4)
sll.insert_at_end(4)
sll.insert_at_end(5)

# Traverse the list before removing duplicates
print("Original List:")
sll.traverse()

# Remove duplicates
sll.remove_duplicates()

# Traverse the list after removing duplicates
print("\nList After Removing Duplicates:")
sll.traverse()
```

```python
#7- Find the intersection of the two linked lists
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    # Function to get the length of the linked list
    def get_length(self):
        length = 0
        current = self.head
        while current:
            length += 1
            current = current.next
        return length

    # Function to find the intersection node
    def get_intersection(self, list2):
        len1 = self.get_length()
        len2 = list2.get_length()

        # Align the starting points
        current1 = self.head
        current2 = list2.head

        # Move the pointer of the longer list ahead
        if len1 > len2:
            for _ in range(len1 - len2):
                current1 = current1.next
        elif len2 > len1:
            for _ in range(len2 - len1):
                current2 = current2.next

        # Move both pointers until they meet
        while current1 and current2:
            if current1 == current2:
                return current1  # Found the intersection node
            current1 = current1.next
            current2 = current2.next

        return None  # No intersection

# List 1: 1 -> 2 -> 3 -> 4 -> 8 -> 6 -> 9
list1 = SinglyLinkedList()
list1.insert_at_end(1)
list1.insert_at_end(2)
list1.insert_at_end(3)
list1.insert_at_end(4)

# Shared part (intersection point)
shared_part = SinglyLinkedList()
shared_part.insert_at_end(6)
shared_part.insert_at_end(9)

# Connect list1's end to the shared part (starting at node 6)
current = list1.head
while current.next:
    current = current.next
current.next = shared_part.head

# List 2: 5 -> 1 -> 6 -> 7
list2 = SinglyLinkedList()
list2.insert_at_end(5)
list2.insert_at_end(1)

# Connect list2's end to the shared part (starting at node 6)
current = list2.head
while current.next:
    current = current.next
current.next = shared_part.head  # This creates the intersection at node 6

# Now list1 and list2 intersect at node 6

# Find intersection
intersection_node = list1.get_intersection(list2)

# Display the result
if intersection_node:
    print(f"Intersection at node with data: {intersection_node.data}")
else:
    print("No intersection")
```

```python
#8- Rotate a linked list by k positions to the right
```

```python
class Node:
    def __init__(self, data):
        self.data = data 
        self.next = None 

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    # Function to traverse the list and print its elements
    def traverse(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    # Function to rotate the linked list by k positions to the right
    def rotate_right(self, k):
        # Edge case: empty list or k is 0
        if not self.head or k == 0:
            return

        # Step 1: Get the length of the list
        current = self.head
        length = 1
        while current.next:
            length += 1
            current = current.next
        
        # Step 2: If k is larger than the length of the list, reduce k
        k = k % length
        if k == 0:
            return

        # Step 3: Make the list circular by connecting the last node to the head
        current.next = self.head

        # Step 4: Find the new tail (length - k - 1) and new head (length - k)
        new_tail = self.head
        for _ in range(length - k - 1):
            new_tail = new_tail.next

        new_head = new_tail.next

        # Step 5: Break the circular link
        new_tail.next = None
        self.head = new_head

# Example Usage:

# Create a linked list
list1 = SinglyLinkedList()

# Insert nodes into the list
list1.insert_at_end(1)
list1.insert_at_end(2)
list1.insert_at_end(3)
list1.insert_at_end(4)
list1.insert_at_end(8)
list1.insert_at_end(6)
list1.insert_at_end(9)

# Print the original list
print("Original List:")
list1.traverse()

# Rotate the list by 2 positions to the right
list1.rotate_right(2)

# Print the rotated list
print("\nList after rotating by 2 positions:")
list1.traverse()
```

```python
#10- Clone a Linked List with next and Random Pointer
```

```python
class Node:
    def __init__(self, data):
        self.data = data  
        self.next = None  
        self.arbit = None 

class SinglyLinkedList:
    def __init__(self):
        self.head = None  

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    # Function to set arbit (random) pointers for the list
    def set_arbit_pointer(self, original_node, arbit_node):
        original_node.arbit = arbit_node

    # Function to traverse the list and print its elements
    def traverse(self):
        current = self.head
        while current:
            arbit = current.arbit.data if current.arbit else None
            print(f"Data: {current.data}, Arbit: {arbit}", end=" -> ")
            current = current.next
        print("None")

    # Function to clone the linked list with both next and arbit pointers
    def clone_list(self):
        if not self.head:
            return None

        # Step 1: Create the clone nodes and insert them next to the original nodes
        current = self.head
        while current:
            new_node = Node(current.data)
            new_node.next = current.next
            current.next = new_node
            current = new_node.next  

        # Step 2: Copy the arbit pointers for the new nodes
        current = self.head
        while current:
            if current.arbit:
                current.next.arbit = current.arbit.next  
            current = current.next.next  

        # Step 3: Separate the cloned list from the original list
        original = self.head
        cloned_head = self.head.next
        cloned_current = cloned_head

        while original:
            original.next = original.next.next
            if cloned_current.next:
                cloned_current.next = cloned_current.next.next  
                cloned_current = cloned_current.next
            original = original.next

        return cloned_head

# Example Usage:

# Create a linked list
list1 = SinglyLinkedList()

# Insert nodes into the list
list1.insert_at_end(1)
list1.insert_at_end(2)
list1.insert_at_end(3)
list1.insert_at_end(4)

# Set arbitrary (random) pointers
list1.set_arbit_pointer(list1.head, list1.head.next.next)  
list1.set_arbit_pointer(list1.head.next, list1.head.next.next.next)
list1.set_arbit_pointer(list1.head.next.next, list1.head)  
list1.set_arbit_pointer(list1.head.next.next.next, list1.head.next)

# Print the original list with arbit pointers
print("Original List with arbit pointers:")
list1.traverse()

# Clone the list
cloned_head = list1.clone_list()

# Print the cloned list with arbit pointers
print("\nCloned List with arbit pointers:")
cloned_list = SinglyLinkedList()
cloned_list.head = cloned_head
cloned_list.traverse()
```
